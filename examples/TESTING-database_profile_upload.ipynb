```python
import cProfile, pstats, io
import datasetdatabase as dsdb

prod = dsdb.DatasetDatabase(config="/active/configs/aics.json", user="jacksonb")
```

```python
import pandas as pd
import numpy as np
np.random.seed(120)

# creating lists of dicts to be formed into a dataframe
ds = []
for i in range(4000):
    d = {}
    d["floats"] = np.random.rand() * 100
    d["ndarrays"] = np.random.rand(2, 2)
    ds.append(d)

# convert this example both to dataframe
ds = pd.DataFrame(ds)
```

```python
ds = dsdb.Dataset(ds)
ds
```

```python
pr = cProfile.Profile()
pr.enable()
ds.upload_to(prod)
pr.disable()
```

```python
s = io.StringIO()
ps = pstats.Stats(pr, stream=s).sort_stats("cumtime")
ps.print_stats()
print(s.getvalue())
```
